```python
from autoscraper import AutoScraper
```

```python
# fetch the link

url = 'https://bj.lianjia.com/ershoufang/'

wanted_dict = {
    'title' : ['融科香雪兰溪  精装复式 业主诚心出售'],
    'totalprice' : ['608万'],
    'address1' : ['融科香雪兰溪'],
    'address2' : ['马驹桥'],
    'type' : ['4室2厅 | 196.99平米 | 南 北 | 简装 | 顶层(共4层) | 2009年建 | 板楼'],
    'unitprice' : ['单价30865元/平米']
}
```

```python
scraper = None
scraper = AutoScraper()
scraper.build(url = url, wanted_dict = wanted_dict)

result = scraper.get_result_similar(url = url, grouped = True)
result
```

```python
url = 'https://bj.lianjia.com/ershoufang/pg2/'

wanted_dict = {
    'title' : ['保利金泉 2室2厅 西北'],
    'totalprice' : ['1050万'],
    'address1' : ['保利金泉'],
    'address2' : ['亚运村'],
    'type' : ['2室2厅 | 114.53平米 | 西北 | 精装 | 中楼层(共30层) | 2008年建 | 塔楼'],
    'unitprice' : ['单价91680元/平米']
}
scraper = None
scraper = AutoScraper()
scraper.build(url = url, wanted_dict = wanted_dict)

result = scraper.get_result_similar(url = url, grouped = True)
result
```

```python
import pandas as pd
```

```python
lianjia_scraper = None
lianjia_scraper = AutoScraper()
lianjia_scraper.load('lianjia-search-20210717')
```

```python
df = pd.DataFrame()

for n in range(10):
    url = f'https://bj.lianjia.com/ershoufang/pg{n}/'
    
    print(url)
    result = lianjia_scraper.get_result_similar(url = url, grouped = True)
    df = pd.concat([df, pd.DataFrame(result)])
```

```python
df.reset_index(drop = True)
```

```python
df.columns = ['title', 'totalprice', 'totalprice2', 'address1', 'address2', 'description', 'unitprice']
df = df[['title', 'totalprice', 'address1', 'address2', 'description', 'unitprice']]
```
